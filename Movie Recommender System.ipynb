```python
import pandas as pd
import numpy as np
import ast
import nltk
```

```python
movies=pd.read_csv("data/tmdb_5000_movies.csv")
credits=pd.read_csv("data/tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies=movies.merge(credits,left_on="id",right_on="movie_id")
```

```python
movies.head(1)
```

```python
#genres
#id
#keywords
#title
#overview
#cast
#crew
movies=movies[["id","genres","keywords","original_title","overview","cast","crew"]]
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
def convert2(obj):
    l=[]
    count=0
    for i in ast.literal_eval(obj):
        if count <= 3:
            l.append(i['name'])
        else:
            break;
    return l
def convert_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
        break;
    return l
```

```python
movies["genres"]=movies["genres"].apply(convert)
movies["keywords"]=movies["keywords"].apply(convert)
#movies["cast"]=movies["cast"].apply(convert)
#movies["crew"]=movies["crew"].apply(convert)
```

```python
movies["cast"]=movies["cast"].apply(convert2)
movies["crew"]=movies["crew"].apply(convert_director)
```

```python
movies.head(1)
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies["genres"]=movies["genres"].apply(lambda x: [i.replace(" ","") for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x: [i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x: [i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies["tags"]=movies["overview"]+movies["genres"]+movies["cast"]+movies["crew"]+movies["keywords"]
```

```python
new_df=movies[["id","original_title","tags"]]
new_df.rename(columns={"original_title":"title"},inplace=True)
```

```python
new_df
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x: " ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem_(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'].apply(stem_)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
cv
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommended(movie):
    movie_index=new_df[new_df["title"]==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        print(i[0])             
```

```python
recommended("Avatar")
```
